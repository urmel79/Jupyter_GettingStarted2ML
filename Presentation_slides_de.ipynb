## STEP 2: Explore the ML dataset

Ziele und Haupt-Arbeitsschritte bei der **Datensatz-Erkundung**:

- Klärung der **Ursprungsgeschichte** des Datensatzes
- Überblick über die interne **Struktur und Organisation** der Daten im Datensatz
- Identifizierung von **Anomalien** im Datensatz
- Vermeidung von **Verzerrungen (Bias)** aufgrund von **Tendenzen**
- Entwicklung erster Ideen bzgl. **innerer Zusammenhänge** (sog. **Korrelationen**) im Datensatz

## STEP 3: Choose and create the ML model

Haupt-Arbeitsschritte bei **Auswahl** und **Erzeugen** des ML-Modells:

- **Kurz-Überblick** über die Welt der KI und **Taxonomie** der ML-Algorithmen
- Auswahl eines geeigneten ML-Algorithmus anhand **Entscheidungsdiagramm**
- Begründung der Wahl des **Support Vector Classifier (SVC)**
- Erläuterung der **Funktionsweise** des SVC
- Erstellen des **SVC-Modells**

## STEP 4: Preprocess the dataset for training

Haupt-Arbeitsschritte beim **Vorbereiten** des Datensatzes:

- Datensatz **bereinigen** (aufgrund Analyse im Schritt 2)
- **Transformation** des Datensatzes durch **Skalierung** der Merkmale durch
    - **Normalisierung** oder
    - **Standardisierung**

## STEP 5: Carry out training, prediction and testing

Haupt-Arbeitsschritte beim **Training** und der **Vorhersage** der Klassen:

- **Aufteilung** des Datensatzes: **Trainings-** und **Testdatensatz**
- **Standardisierung** der Datensätze
- **Trainieren** des SVC-Modells
- **Vorhersagen** treffen mithilfe des ungelabelten **Testdatensatzes**

## STEP 6: Evaluate model's performance

Bewertung der **Qualität des Klassifikationsergebnisses** anhand bekannter **Metriken**:

- Genauigkeitswert (**Accuracy Score**)
- Klassifikationsbericht (**Classification Report**)
- Kreuzvalidierungsergebnis (**Cross-validation score**)
- Konfusionsmatrix (**Confusion matrix**)

## STEP 7: Vary parameters of the ML model manually

Schritte beim **manuellen Variieren** der **Hyperparameter** des SVC:

- Vorbereitung:
    - Datensatz aufbereiten
    - Plotting-Funktionen für vergleichende Darstellung erstellen
- Paramater-Variation:
    - **Kernel** des SVC variieren
    - **Gamma**-Parameter variieren
    - **C**-Parameter variieren
    - Variation des **Grades** (Ordnung)

Ziel: Demonstration des **Einflusses** auf das **Klassifikationsergebnis** durch **manuelle Variation** der einzelnen **Hyperparameter**

## STEP 8: Tune the ML model systematically

Schritte beim **systematischen Variieren** der **Hyperparameter** des SVC:

- Ermittlung einer **Basislinie** (SVC mit default Hyperparametern)
- Hinzufügen von **Gaußschem Rauschen** zum Iris-Datensatz (**Messrauschen** simulieren)
- **Rastersuche** durchführen
- **Zufallssuche** ausführen

Ziel: **Automatische Suche** nach günstigen **Hyperparametern** &rarr; Optimieren des **Klassifikationsergebnisses**
